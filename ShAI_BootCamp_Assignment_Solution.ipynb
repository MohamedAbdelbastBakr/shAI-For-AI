```python
import pandas as pd
import numpy as np

# Load your dataset
df = pd.read_csv('/content/drive/MyDrive/Salaries.csv')
df.head()
```

**Basic Data Exploration**


```python
print(df.head(5))
print(df.info)
num_rows, num_columns = df.shape
# Identify the number of rows and columns
print(f"\nNumber of rows: {num_rows}")
print(f"Number of columns: {num_columns}")
# Data types of each column
print(f"\nData types of each column:\n{df.dtypes}")
# Check for missing values
print(f"\nMissing values in each column:\n{df.isnull().sum()}")

```

**Descriptive Statistics**

```python
# Calculate basic statistics
mean_salary = df['TotalPayBenefits'].mean()
print(f"Mean salary: {mean_salary:.3f}")
median_salary = df['TotalPayBenefits'].median()
print(f"Median salary: {median_salary:.3f}")
mode_salary = df['TotalPayBenefits'].mode().iloc[0]
print(f"Mode salary: {mode_salary:.3f}")
min_salary = df['TotalPayBenefits'].min()
print(f"Minimum salary: {min_salary:.3f}")
max_salary = df['TotalPayBenefits'].max()
print(f"Maximum salary: {max_salary:.3f}")
print(f"Salary range: from {min_salary} to {max_salary:.3f}")
std_dev_salary = df['TotalPayBenefits'].std()
print(f"Standard deviation of salary: {std_dev_salary:.3f}")
```

**Data Cleaning**

```python
ddf = pd.read_csv('/content/drive/MyDrive/Salaries.csv')
print(f"Missing values before imputation:\n{ddf.isnull().sum()}")

'''I'll demonstrate how to impute missing values in the columns with the mean
value.Imputing with the mean is a simple approach that can work well when the
missing values are randomly distributed and the mean is a representativevalue
for the column.if missing values are categorical columns,it might be more
appropriate to impute missing values with a placeholder category or label.'''

df['OvertimePay'].fillna(df['OvertimePay'].mean(), inplace=True)
df['OtherPay'].fillna(df['OtherPay'].mean(), inplace=True)
df['Benefits'].fillna(df['Benefits'].mean(), inplace=True)
df['Notes'].fillna('No Notes', inplace=True)
df['Status'].fillna('un known status', inplace=True)
df['TotalPayBenefits'] = df['TotalPayBenefits'].apply(lambda x: x if x >= 0 else pd.NaT)
mean_TotalPayBenefits = df['TotalPayBenefits'].mean()
df['TotalPayBenefits'].fillna(mean_TotalPayBenefits, inplace=True)

# Display the number of missing values after imputation
print("\nMissing values after imputation:")
print(df.isnull().sum())

mean_salary = df['TotalPayBenefits'].mean()
print(f"Mean salary: {mean_salary:.3f}")
median_salary = df['TotalPayBenefits'].median()
print(f"Median salary: {median_salary:.3f}")
mode_salary = df['TotalPayBenefits'].mode().iloc[0]
print(f"Mode salary: {mode_salary:.3f}")
min_salary = df['TotalPayBenefits'].min()
print(f"Minimum salary: {min_salary:.3f}")
max_salary = df['TotalPayBenefits'].max()
print(f"Maximum salary: {max_salary:.3f}")
print(f"Salary range: from {min_salary} to {max_salary:.3f}")
std_dev_salary = df['TotalPayBenefits'].std()
print(f"Standard deviation of salary: {std_dev_salary:.3f}")
```

**Basic Data Visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn plots
sns.set(style="whitegrid")

# Histogram to visualize the distribution of salaries
plt.figure(figsize=(10, 6))
sns.histplot(df['TotalPayBenefits'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Salaries')
plt.xlabel('Total Pay + Benefits')
plt.ylabel('Frequency')
plt.show()

# Pie chart to represent the proportion of employees in different departments
plt.figure(figsize=(10, 6))
agency_counts = df['JobTitle'].value_counts()
plt.pie(agency_counts, labels=agency_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

**Grouped Analysis**

```python
grouped_data = df.groupby('JobTitle')['TotalPayBenefits'].agg(['count', 'mean', 'median', 'min', 'max', 'std'])
print(grouped_data)
sorted_data = grouped_data.sort_values(by='mean', ascending=False)
print(sorted_data)
```

**Simple Correlation Analysis**

```python
import matplotlib.pyplot as plt
import seaborn as sns

correlation_1 = df['TotalPayBenefits'].corr(df['BasePay'])
print(f"Correlation between TotalPayBenefits and BasePay: {correlation_1:.2f}")
plt.figure(figsize=(10, 6))
sns.scatterplot(x='BasePay', y='TotalPayBenefits', data=df, color='skyblue')
plt.title('Scatter Plot: TotalPayBenefits vs. BasePay')
plt.xlabel('BasePay')
plt.ylabel('Salary')
plt.show()

correlation_2 = df['TotalPayBenefits'].corr(df['OvertimePay'])
print(f"Correlation between TotalPayBenefits and OvertimePay: {correlation_2:.2f}")
plt.figure(figsize=(10, 6))
sns.scatterplot(x='OvertimePay', y='TotalPayBenefits', data=df, color='skyblue')
plt.title('Scatter Plot: TotalPayBenefits vs. OvertimePay')
plt.xlabel('OvertimePay')
plt.ylabel('Salary')
plt.show()

correlation_3 = df['TotalPayBenefits'].corr(df['OtherPay'])
print(f"Correlation between TotalPayBenefits and OtherPay: {correlation_3:.2f}")
plt.figure(figsize=(10, 6))
sns.scatterplot(x='OtherPay', y='TotalPayBenefits', data=df, color='skyblue')
plt.title('Scatter Plot: TotalPayBenefits vs. OtherPay')
plt.xlabel('OtherPay')
plt.ylabel('Salary')
plt.show()

correlation_4 = df['TotalPayBenefits'].corr(df['Benefits'])
print(f"Correlation between TotalPayBenefits and Benefits: {correlation_4:.2f}")
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Benefits', y='TotalPayBenefits', data=df, color='skyblue')
plt.title('Scatter Plot: TotalPayBenefits vs. Benefits')
plt.xlabel('Benefits')
plt.ylabel('Salary')
plt.show()

correlation_5 = df['TotalPayBenefits'].corr(df['TotalPay'])
print(f"Correlation between TotalPayBenefits and TotalPay: {correlation_5:.2f}")
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TotalPay', y='TotalPayBenefits', data=df, color='skyblue')
plt.title('Scatter Plot: TotalPayBenefits vs. TotalPay')
plt.xlabel('TotalPay')
plt.ylabel('Salary')
plt.show()

correlation_6 = df['TotalPayBenefits'].corr(df['Year'])
print(f"Correlation between TotalPayBenefits and Year: {correlation_6:.2f}")
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Year', y='TotalPayBenefits', data=df, color='skyblue')
plt.title('Scatter Plot: TotalPayBenefits vs. Year')
plt.xlabel('Year')
plt.ylabel('Salary')
plt.show()
```
