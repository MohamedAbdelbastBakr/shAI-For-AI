### Import The Libraries And Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('heart.csv')
```

### dataset content

*age

*sex

*chest pain'cp'

    -0--> typical angina
    -1--> atypical angina
    -2--> non-anginal pain
    -3--> asymptomatic
*trestbps: reading blood pressure (in mm Hg on admission to the hospital)

*chol:serum cholestoral in mg/dl

*fbs:(fasting blood sugar >120 mg/dl)(1-->true , 0-->false)

*restecg: resting electrocardiographic results

    -Value 0: normal
    -Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
    -Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria

*thalach: maximum heart rate achieved

*exang: exercise (riduced angina (1 = yes; 0 = no)

*oldpeak = ST depression induced by exercise relative to rest

*slope: the slope of the peak exercise ST segment

    -Value 1: upsloping
    -Value 2: flat
    -Value 3: downsloping

*ca: number of major vessels (0-3) colored by flourosopy

*thal: 3= normal; 6 = fixed defect; 7 = reversable defect

*target: 0=less chance of heart attack, 1=more chance of heart attack

### How Many People Have less and more chance of heart attack?

```python
# Count the occurrences of each class in the 'target' column
target_counts = data['target'].value_counts()

# Plot a bar plot showing the count of each target class
plt.bar(target_counts.index.astype(str), target_counts.values, color=['blue', 'orange'])

# Add labels and title to the plot
plt.xlabel('Target')
plt.ylabel('Count')
plt.title('Count of Each Target Class')
plt.xticks([0,1],['More Chance','Less Chance'])

# Add annotations for each bar
for i, count in enumerate(target_counts):
    plt.text(i, count, str(count), ha='center', va='bottom')

# Show the plot
plt.show()
```

### Plot Continuous Variables

```python
data.columns
```

separate columns which contain continuous values and columns which contain categorical values

```python
cont_val=[]
cate_val=[]

for column in data.columns:
    if data[column].nunique() <=10:
        cate_val.append(column)
    else:
        cont_val.append(column)
```
